## Prepare - Trips Data

As the data is stored in a number of different .csv files, we do need to prepare the data before we can start process/clean it.
In this python worksheet, we aim to achieve the following:
1. Standarize the column names for all the sheets. 
2. Combine all the trip data into one .csv file.

```python
# Installing required libraries and setting the panda's display options.
import numpy as np
import pandas as pd
import os
import gc
import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_colwidth', 999)
pd.set_option('display.max_rows', 9999)
```

```python
# We will begin by creating a pandas dataframe to store the file name, column names in a list and record numbers.
# This is just to look at the column names across files.
path = r"C:\Users\utsav\OneDrive\Documents\ML\Google Data Analytics Course\Course 8 - Capstone\Case Study 1\data\Unzipped\Combined\trips"
os.chdir(path)
df = pd.DataFrame(columns = ['file_name', 'col_names', 'num_records'])
```

```python
files = 0

for file in os.listdir():
    file_path = f"{path}\{file}"
    with open(file_path, 'r') as f:
        print(f'File Number Opening: {files+1}')
        cols = f.readline()
        cols = cols[:-2]
        num_records = sum(1 for line in f) - 1
        cols_list = cols.split(',')
        df.at[files,'file_name'] = file
        df.at[files,'col_names'] = cols_list
        df.at[files,'num_records'] = num_records
        files += 1
```

```python
del df
gc.collect()
```

```python
'''
Our first task is to standerize the column names. 
We will use "202105-divvy-tripdata.csv" column names. However, because of large number of data points, we were not able
to open the same in the spreadsheets, like we did for station data. 
'''
files = 0
trip_list = []
for f in os.listdir():
    print(f'File Number Opening: {files+1}')
    trip_df = pd.read_csv(f, sep=',')
    trip_df['file'] = f.split('/')[-1]
    trip_list.append(trip_df)
    files += 1
    
trips_combined_data = pd.concat(trip_list, ignore_index=True, sort=True)
```

```python
# List of unique names of columns in all the data sheets combined.
trips_combined_data.info()
```

```python
# Creating a dictionary to link all the related column names to the startarized one.
columns_dict ={'ride_id': ['trip_id','01 - Rental Details Rental ID'],
              'started_at': ['01 - Rental Details Local Start Time','start_time','starttime'],
              'ended_at': ['01 - Rental Details Local End Time', 'stoptime','end_time'],
               'start_station_name': ['03 - Rental Start Station Name', 'from_station_name'],
               'start_station_id': ['03 - Rental Start Station ID','from_station_id'],
               'end_station_name': ['02 - Rental End Station Name', 'to_station_name'],
               'end_station_id': ['02 - Rental End Station ID','to_station_id'],
               'birthyear': ['05 - Member Details Member Birthday Year','birthday'],
               'bikeid': ['01 - Rental Details Bike ID'],
               'usertype': ['User Type', 'member_casual'],
               'gender' : ['Member Gender'],
               'tripduration' : ['01 - Rental Details Duration In Seconds Uncapped']           
              }
```

```python
for key, value in columns_dict.items():
    for item in value:
        trips_combined_data[key].fillna(trips_combined_data[item], inplace = True)
        trips_combined_data.drop([item], axis = 1, inplace = True)
        gc.collect()
    print(key, trips_combined_data[key].isna().sum())

trips_combined_data.info()
```

```python
# After the above step, we are left with a much smaller and standerized columns names.
trips_combined_data.sample(25)
```

```python
# Due to combining the data from a multiple files, we would need to standerize the column contents as well.
# Though, we can do the same in the process part of our project, I chose to do the steps here, along with combining data and standerizing.
trips_combined_data['usertype'].value_counts()
```

```python
trips_combined_data.loc[trips_combined_data['usertype'] == 'Subscriber','usertype'] = "member"
trips_combined_data.loc[trips_combined_data['usertype'] == 'Dependent','usertype'] = "member"
trips_combined_data.loc[trips_combined_data['usertype'] == 'Customer','usertype'] = "casual"
trips_combined_data['usertype'].value_counts()
```

```python
trips_combined_data.columns
```

```python
# Organinzing the data. 
cols = ['ride_id', 'started_at', 'start_station_id', 'start_station_name', 'start_lat', 'start_lng',
        'ended_at', 'end_station_id', 'end_station_name', 'end_lat', 'end_lng',
       'bikeid', 'birthyear',
        'usertype', 'tripduration', 'gender', 'rideable_type', 'file']
trips_combined_data = trips_combined_data[cols]
decimals = pd.Series([5, 5, 5, 5], index=['start_lat', 'start_lng','end_lat','end_lng'])
trips_combined_data = trips_combined_data.round(decimals)
trips_combined_data.sort_values(by = "started_at", inplace = True)
```
